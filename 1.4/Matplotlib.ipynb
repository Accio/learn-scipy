# A simple plot

Plot with default setting

## the default setting

```python
import numpy as np

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)


plt.plot(X, C)
plt.plot(X, S)

plt.show()
```

## Understanding the defaults with instances

```python
# Create a figure of size 4x3 inches, 300 dots per inch
plt.figure(figsize=(4, 3), dpi=300)

# Create a new subplot from a grid of 1x1
plt.subplot(1, 1, 1)

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

# Plot cosine with a blue continuous line of width 1.25 (pixels)
plt.plot(X, C, color="blue", linewidth=1.25, linestyle="-")

# Plot sine with a green continuous line of width 1.25 (pixels)
plt.plot(X, S, color="green", linewidth=1.25, linestyle="-")

# Set x limits
plt.xlim(-4.0, 4.0)

# Set x ticks
# plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])

# Set y limits
plt.ylim(-1.0, 1.0)

# Set y ticks
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))

# Set x/y lab
plt.xlabel("Input")
plt.ylabel("Transcendent")

# Save figure using 72 dots per inch
plt.savefig("exercice_2.png", dpi=300)

# Show result on screen
plt.show()
```
